```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import datetime as dt
import plotly.express as px
```

```python
df=pd.read_csv('covid_19_data.csv')
df
```

```python
df['ObservationDate']=pd.to_datetime(df['ObservationDate'])
df['Last Update']=pd.to_datetime(df['Last Update'])
df
```

```python
df.groupby(by='Country/Region').sum()['Confirmed'].sort_values(ascending=False)
```

```python
print('Total No. of Confirmed cases are' ,df['Confirmed'].sum())
```

```python
sns.pairplot(data=df)
```

```python
plt.figure(figsize=(15,6))
sns.violinplot(x='Country/Region',y='Confirmed',data=df)
plt.xticks(rotation=90)
```

```python
df1=[df.Confirmed.sum(),df.Deaths.sum(),df.Recovered.sum()]
plt.pie(x=df1,labels=['Confirmed','Deaths','Recovered'],autopct='%.2f%%')
```

```python
df[['Confirmed','Deaths','Recovered']].sum().plot(kind="pie",autopct='%.2f%%')
```

```python
plt.figure(figsize=(20,8))
df.groupby(by='Country/Region').sum()['Confirmed'].sort_values(ascending=False).plot(kind='bar')
```

```python
df['Month']=df.ObservationDate.dt.month
df.groupby(by=['Month','Country/Region'])['Confirmed'].sum().plot(kind='bar')
```

```python
df.groupby(by='Country/Region').mean()['Deaths'].sort_values(ascending=False).plot(kind='bar')
```
