Q. 1) Instruction ( For Data Cleaning ) - Find all Null Values in the dataset. If there is any null value in any column, then fill it with the mean of that column.

```python
val = {}
for i in df.columns:
    if df[i].dtypes == "O":
        val[i]=df[i].mode()[0]
    else:
        val[i]=df[i].mean()

val

df.fillna(value=val,inplace=True)
```

```python
li = []
for i in df.MSRP.values:
    li.append(float(i[1:].replace(",","")))
li
```

```python
df["RMSRP"] = li
```

```python
df.isnull().sum()
```

```python
df['Rinvoice']=df.Invoice.map(lambda x: float(x[1:].replace(',','')))
df
```

Q. 4) Instruction ( Removing unwanted records ) - Remove all the records (rows) where Weight is above 4000.

```python
df.drop(index=(df[df.Weight > 4000].index),inplace=True)
```

Q. 5) Instruction ( Applying function on a column ) - Increase all the values of 'MPG_City' column by 3.

```python
df.MPG_City = df.MPG_City.map(lambda x: x +3)
```
