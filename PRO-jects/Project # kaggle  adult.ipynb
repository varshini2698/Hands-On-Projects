```python
# Enter your code here. Read input from STDIN. Print output to STDOUT
#a = input()

a = "Swapnil1234"
ints = []
alphs = []
for i in a:
    if i.isdigit():
        ints.append(i)
    else:
        alphs.append(i)
oddd = []
evnn = []
for i in ints:
    if int(i)%2==0:
        evnn.append(i)
    else:
        oddd.append(i)

upp = []
lww = []

for i in alphs:
    if i.isupper() == True:
        upp.append(i)
    else:
        lww.append(i)
        
lww.sort()
upp.sort()
evnn.sort()
oddd.sort()

b = (lww+upp+oddd+evnn)
for i in b:
    print(i,end="")
```
